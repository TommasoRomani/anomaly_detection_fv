```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt

NUM_FEATURES=20
```

```python
df = pd.read_csv("./data/merged_annuale_meteo_sunrise_filtrato15_comm_code.csv", delimiter=";")

tmp = []
for i in df.index:
    if df["INV02_CommunicationCode_60"].loc[i]==1 or df["INV03_CommunicationCode_60"].loc[i]==1 or df["Cont_CommunicationCode_60"].loc[i]==1:
        tmp.append(1)
    else:
        tmp.append(0)
        
    
```

```python
df["true_anomalies"]=tmp
```

```python

train_df = pd.read_csv("./results/res_train.csv", delimiter=";")
test_df = pd.read_csv("./results/res_test.csv", delimiter=";")



df_train = df.drop(index=df.index[:100], axis=0)
df_train.drop(index=df_train.index[28604:], axis=0, inplace=True)
df_test = df.drop(index=df.index[:28804], axis=0)

for col in range(NUM_FEATURES):
    tmp = []
    for i in train_df.index:
        if train_df[f"A_Pred_{col}"].loc[i]==1 or train_df["A_Pred_Global"].loc[i]==1:
            tmp.append(1)
        else:
            tmp.append(0)
    train_df[f"pred_{col}"] = tmp
```

```python
def compute_metrics(y_true, y_pred):
    
    metrics={
    'Accuracy': accuracy_score(y_true, y_pred),
    "Precision": precision_score(y_true, y_pred),
    "Recall": recall_score(y_true, y_pred),
    "F1 Score": f1_score(y_true, y_pred),
    }
    '''
    metrics['Accuracy'] = accuracy
    metrics["Precision"] = precision
    metrics['Recall'] = recall
    metrics['F1 Score'] = f1
    metrics['Confusion Matrix'] = conf_matrix
    '''
    return metrics
```

```python
metrics = {}
for i in range(NUM_FEATURES):
    metrics[f"feature{i}"] = compute_metrics(df_train["true_anomalies"], train_df[f"pred_{i}"])
```

```python
metric_df = pd.DataFrame.from_dict(metrics)
```

```python
def compute_ROC(y_true, y_scores, i):
    fpr, tpr, thresholds = roc_curve(y_true, y_scores)
    roc_auc = auc(fpr, tpr)
    # Plot ROC curve
    plt.figure(figsize=(8, 8))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    
    plt.savefig(f'./graphs/ROC_Curves/AUC_ROC_{i}.png')
```

```python
for i in range(NUM_FEATURES):
    compute_ROC(df_train["true_anomalies"], train_df[f"A_Score_{i}"], i)
```
