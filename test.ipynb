```python
for col in df:
    if any(key in col for key in "INV"):
        for i,val in enumerate(df[col].values):
            if val == 60:
                n_df[f"{col}_60"].loc[i] = 60
            if val == 116:
                n_df[f"{col}_116"].loc[i] = 116
            
    elif col=="Cont_CommunicationCode":
        for i,val in enumerate(df[col].values):
            if val == 60:
                n_df[f"{col}_60"].loc[i] = 60
            if val == 118:
                n_df[f"{col}_118"].loc[i] = 118
```
